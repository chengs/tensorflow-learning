```python
%matplotlib inline

import tensorflow as tf
from tensorflow import keras
```

```python
inputs = keras.Input(shape=(10,))
layer1 = keras.layers.Dense(48, activation='relu', input_shape=(10,))(inputs)
layer21 = keras.layers.Dense(20, activation='tanh')(layer1)
layer22 = keras.layers.Dense(10, activation='sigmoid')(layer1)
layer3 = keras.layers.Concatenate()([layer21, layer22])
outputs = keras.layers.Dense(4, activation='softmax')(layer3)

model = keras.Model(inputs=inputs, outputs=outputs)
model.summary()
```

```python
keras.utils.plot_model(model,show_shapes=True)
```

```python
# normal save
model.save('model_keras_save')
```

```python
# tf save
model.save('./model_tf_save/', save_format='tf')
```

```python
# h5 save
model.save('model_h5_save', save_format='h5')
```
